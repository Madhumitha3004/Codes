```python
# aggregate, filter, transform, apply
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'], 'data1': range(6), 'data2': rng.randint(0, 10, 6)}, columns = ['key', 'data1', 'data2'])
df
```

```python
# aggregation
df.groupby('key').aggregate(['min',np.median, max])
```

```python
df.groupby('key').aggregate({'data1': 'min','data2': 'max'})
```

```python
# filtering
#A filtering operation allows to drop data based on the group properties. For example, to keep all groups in which the standard deviation is larger than some critical value:
def filter_func(x):
    return x['data2'].std() > 4

display('df', "df.groupby('key').std()", "df.groupby('key').filter(filter_func)")
```

```python
# transformation : transformation can return some transformed version of the full data to recombine
df.groupby('key').transform(lambda x: x - x.mean())
```

```python
# apply
def norm_by_data2(x):
    # x is a DataFrame of group values
    x['data1'] /= x['data2'].sum()
    return x

display('df', "df.groupby('key').apply(norm_by_data2)")
```
